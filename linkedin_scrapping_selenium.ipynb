```python
#import libraries
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import undetected_chromedriver.v2 as uc
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#setting up chrome driver
driver = uc.Chrome(use_subprocess=True)
driver.get('https://www.linkedin.com/') #opening page in chrome
#initialising variables
profile_info=[]
profile_link=[]
first_name=[]
last_name=[]
profile_title=[]
location=[]
while 1:
    try:
        people_button = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,"/html/body/nav/ul/li[2]/a")))   
        people_button.click() 
        firstname_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"input[placeholder='First Name']")))
        lastname_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"input[placeholder='Last Name']")))
        break
    except:
        try:
            driver.refresh()
            people_button = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH,"/html/body/nav/ul/li[2]/a")))   
            people_button.click() 
            firstname_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"input[placeholder='First Name']")))
            lastname_input = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR,"input[placeholder='Last Name']")))
            break
        except:
            driver.close()
            driver = uc.Chrome(use_subprocess=True)
            driver.get('https://www.linkedin.com/')
            pass
try:
    
    firstname_input.send_keys("Firstname") #input firstname as <Firstname>
    lastname_input.send_keys("Lastname") #input larstname as <Lastname>
    driver.find_element(by=By.CLASS_NAME,value="base-search-bar__submit-btn").click()
    profiles = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.CLASS_NAME,"pserp-layout__profile-result-list-item")))    
    
    
    for i in range(10):  #initialising loop for first 10 profiles
        profile = WebDriverWait(profiles[i], 10).until(EC.presence_of_element_located((By.CLASS_NAME,"base-card")))   
        profile.send_keys(Keys.CONTROL + Keys.RETURN) #opening profile in new tab
        driver.switch_to.window(driver.window_handles[1]) # switching to new tab
        try:
            element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[@id="public_profile_contextual-sign-in"]/div/section/button')))
            driver.find_element(by=By.XPATH,value='//*[@id="public_profile_contextual-sign-in"]/div/section/button').click()    
        except:
            pass
        profile_link.append(driver.current_url) 
        profile_info.append(driver.page_source) #Storing profile html in the list
        driver.close()
        driver.switch_to.window(driver.window_handles[0])      
    
except Exception as e:
    print("the code faced following exception :"," ",type(e).__name__)
    pass
driver.close()
# Retriving information from profile pages by parsing html using Beautifulsoup

for info in profile_info:
    soup = BeautifulSoup(info,'html.parser')
    name= soup.find("h1",class_="top-card-layout__title").text
    firstname,lastname = name.replace("\n","").strip().split()
    title=soup.find("h2",class_="top-card-layout__headline").text.replace("\n","").strip()
    loc=soup.find("div",class_="top-card__subline-item").text.strip()
    first_name.append(firstname)
    last_name.append(lastname)
    location.append(loc)
    profile_title.append(title)
    
    
df = pd.DataFrame() #creating a new DataFrame to store the information of profiles
df["firstname"] = first_name
df["lastname"] = last_name
df["title"] = profile_title
df["location"] = location
df["profile_link"] = profile_link
#Saving dataframe as linkedin_info.csv
df.to_csv("linkedin_info.csv")
```
